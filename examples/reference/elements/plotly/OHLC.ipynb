#### **Title**: OHLC Element

**Dependencies**: Plotly

**Backends**: [Bokeh](../bokeh/OHLC.ipynb), [Matplotlib](../matplotlib/OHLC.ipynb), [Plotly](./OHLC.ipynb)

```python
import numpy as np
import pandas as pd

import holoviews as hv

hv.extension('plotly')
```

The ``OHLC`` Element draws a candlestick chart of the open, high, low and close prices of a financial instrument over a sequence of timestamps. The key dimension maps the timestamps (or any numeric coordinate) onto the x-axis, while the four value dimensions define, in order, the ``open``, ``high``, ``low`` and ``close`` values.

Each candle is composed of a *body* spanning the open and close values and a *wick* spanning the low and high values. The body color encodes the direction: a close at or above the open is drawn as an "up" candle, while a close below the open is drawn as a "down" candle. The Plotly backend renders this with a native candlestick trace.

The simplest way to construct an ``OHLC`` Element is from a list of ``(x, open, high, low, close)`` tuples:

```python
data = [
    (0, 100, 105, 99, 104),
    (1, 104, 107, 103, 103.5),
    (2, 103.5, 108, 102, 107),
    (3, 107, 109, 105, 105.5),
    (4, 105.5, 106, 101, 102),
]
hv.OHLC(data)
```

More commonly the data lives in a dataframe indexed by a datetime column. The key dimension is supplied explicitly, followed by the four value dimensions in open, high, low, close order:

```python
rng = np.random.default_rng(42)
dates = pd.date_range('2024-01-01', periods=40, freq='D')
close = 100 + rng.standard_normal(40).cumsum()
open_ = close + rng.standard_normal(40)
high = np.maximum(open_, close) + rng.random(40)
low = np.minimum(open_, close) - rng.random(40)
df = pd.DataFrame({'date': dates, 'open': open_, 'high': high, 'low': low, 'close': close})

hv.OHLC(df, 'date', ['open', 'high', 'low', 'close']).opts(width=700, xlabel='Date', ylabel='Price')
```

The up and down body colors can be customized with the ``pos_color`` and ``neg_color`` options:

```python
hv.OHLC(df, 'date', ['open', 'high', 'low', 'close']).opts(
    width=700, pos_color='#2ca02c', neg_color='#d62728'
)
```

Any value dimensions beyond the first four are retained on the Element.

For full documentation and the available style and plot options, use ``hv.help(hv.OHLC)``.
